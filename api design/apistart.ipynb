# FastAPI Quick Start

This notebook gives you a **simple FastAPI endpoint** and a short guide to run it.

```python
# 1) Install dependencies (run once)
# !pip install fastapi uvicorn

from fastapi import FastAPI

app = FastAPI(title="Simple API")

@app.get("/")
def read_root():
    return {"message": "Hello from FastAPI"}

@app.get("/health")
def health():
    return {"status": "ok"}

print("App created. Save this code into main.py to run with uvicorn.")
```

## Run Guide

1. Create a file named `main.py` and paste the app code from the cell above.
2. In terminal, install dependencies:
   - `pip install fastapi uvicorn`
3. Start server:
   - `uvicorn main:app --reload`
4. Open in browser:
   - API root: `http://127.0.0.1:8000/`
   - Health check: `http://127.0.0.1:8000/health`
   - Swagger docs: `http://127.0.0.1:8000/docs`

Quick test in terminal:
- `curl http://127.0.0.1:8000/health`
