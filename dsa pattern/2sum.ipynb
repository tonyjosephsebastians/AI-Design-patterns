Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
Example 2:

Input: nums = [3,2,4], target = 6
Output: [1,2]
Example 3:

Input: nums = [3,3], target = 6
Output: [0,1]
 

Constraints:

2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109

```python
class Solution:

    def sumofval(self,nums ,target):
        if len(nums) < 2:
            return []
        arr = sorted((num, i) for i, num in enumerate(nums))

        left,right = 0,len(nums)-1


        while(left < right):
            total = arr[left][0] + arr[right][0]
            if total == target:
                return [arr[left][1], arr[right][1]] 

            elif total > target:
                right -=1
            else :
                left +=1
        
        return []

            
```

```python
class Solution:
    def sumofval(self, nums, target, debug=True):
        if len(nums) < 2:
            if debug:
                print("Array has fewer than 2 elements.")
            return []

        # Keep original index while sorting by value
        arr = sorted((num, i) for i, num in enumerate(nums))
        left, right = 0, len(arr) - 1

        if debug:
            print(f"Input nums: {nums}")
            print(f"Target: {target}")
            print(f"Sorted (value, original_index): {arr}")

        step = 1
        while left < right:
            lv, li = arr[left]
            rv, ri = arr[right]
            total = lv + rv

            if debug:
                print(
                    f"Step {step}: "
                    f"left={left} -> ({lv}, idx={li}), "
                    f"right={right} -> ({rv}, idx={ri}), "
                    f"sum={total}"
                )

            if total == target:
                if debug:
                    print(f"Found! Returning original indices: [{li}, {ri}]")
                return [li, ri]
            elif total < target:
                if debug:
                    print("sum < target, move left pointer right")
                left += 1
            else:
                if debug:
                    print("sum > target, move right pointer left")
                right -= 1

            step += 1

        if debug:
            print("No pair found.")
        return []
```

```python
Sol = Solution()
nums = [3,3]
target = 6

print(Sol.sumofval(nums,target))
```

```python
class Solution:
    def sumofval(self, nums, target):
        seen = {}  # value -> index
        for i, n in enumerate(nums):
            need = target - n
            if need in seen:
                return [seen[need], i]
            seen[n] = i
        return []
```

```python
Sol = Solution()
nums = [3,3]
target = 6

print(Sol.sumofval(nums,target))
```

For the standard Two Sum on an unsorted array (return indices), hashmap is best.

Hashmap: O(n) time, O(n) space
Two-pointer (with sorting): O(n log n) time, O(n) space (or O(1) extra if in-place sort, but index handling gets tricky)
Use two-pointer mainly when:

the array is already sorted, or
you care about pair values more than original indices.
