Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.

 

Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation: 
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
Notice that the order of the output and the order of the triplets does not matter.
Example 2:

Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.
Example 3:

Input: nums = [0,0,0]
Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.

```python
class Solution:
    def sumnums(self,nums:List[int]):
        if len(nums) <1:
            return []

        left,right = 0,2
        result = []

        while left < right and right <=len(nums):
            if sum(nums[left:right+1]) ==0:
                result.append(list((nums[left:right+1])))
            else :
                left +=1
                right +=1

        return result


            




```

```python
from typing import List

class Solution:
    def sumnums(self, nums: List[int]) -> List[List[int]]:
        if len(nums) < 3:
            return []

        nums.sort()
        print(f"Sorted nums: {nums}")

        result = []

        for i in range(len(nums) - 2):
            # skip duplicate 'i' values
            if i > 0 and nums[i] == nums[i - 1]:
                print(f"Skip i={i} because nums[{i}] == nums[{i-1}] == {nums[i]}")
                continue

            left, right = i + 1, len(nums) - 1
            print(f"\nFix i={i}, nums[i]={nums[i]}, start left={left}, right={right}")

            while left < right:
                total = nums[i] + nums[left] + nums[right]
                print(
                    f"Check -> i={i}({nums[i]}), left={left}({nums[left]}), "
                    f"right={right}({nums[right]}), total={total}"
                )

                if total == 0:
                    triplet = [nums[i], nums[left], nums[right]]
                    result.append(triplet)
                    print(f"  Found triplet: {triplet}")

                    left += 1
                    right -= 1

                    # skip duplicates for left
                    while left < right and nums[left] == nums[left - 1]:
                        print(f"  Skip duplicate left at index {left}, value={nums[left]}")
                        left += 1

                    # skip duplicates for right
                    while left < right and nums[right] == nums[right + 1]:
                        print(f"  Skip duplicate right at index {right}, value={nums[right]}")
                        right -= 1

                elif total < 0:
                    print("  Total < 0, move left to increase sum")
                    left += 1
                else:
                    print("  Total > 0, move right to decrease sum")
                    right -= 1

        print(f"\nFinal result: {result}")
        return result


# ----- sample call -----
sol = Solution()
sample_nums = [-1, 0, 1, 2, -1, -4]
answer = sol.sumnums(sample_nums)
print("Returned:", answer)
```
